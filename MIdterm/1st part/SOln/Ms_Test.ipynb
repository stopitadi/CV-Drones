```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
    img = cv2.imread("/Users/akashpaijwar/Documents/Winter Project/CV101/MS/1st part/SOln/cherrytree.jpeg")
    
    # Display the original image
    #plt.imshow(img[:,:,::-1])
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    plt.axis('off')
    plt.show()
```

```python
    img_bright = cv2.convertScaleAbs(img, alpha=0.5, beta=0)

    # Display the image after adjusting brightness
    plt.imshow(cv2.cvtColor(img_bright, cv2.COLOR_BGR2RGB))
    plt.title("Brightness Adjusted")
    plt.axis('off')
    plt.show()
```

```python
img_contrast = cv2.convertScaleAbs(img_bright, alpha=1.5, beta=0)
plt.imshow(img_contrast[:, :, ::-1])  # Reversing color channels from BGR to RGB
plt.title("Contrast Adjusted")
plt.axis('off')
plt.show()
```

```python
img_rgb = cv2.cvtColor(img_contrast, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
hsv[:, :, 1] = np.clip(hsv[:, :, 1] * 1.5, 0, 255)
img_saturation = cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)

# Display the final filtered image using Matplotlib
plt.imshow(img_saturation)
plt.title("Final Filtered Image")
plt.axis('off')
plt.show()
```
